```python
from foreground_sim_fast import ForeGsim
import astropy.io.fits as fits
import numpy as np
import matplotlib.pyplot as plt
```

```python
sim_test = ForeGsim(nfreq=316 ,numin = 101.14547051514067, nustep = 0.17978510057201902,ngrid = 150, imgsize=3.1362739101377324)
```

```python
snr_test = sim_test.gen_snr_map()
```

```python
plt.imshow(snr_test[:,:,0])

print(np.unique(snr_test))
```

```python

syn_test = sim_test.gen_sync_map()
'''hdu = fits.PrimaryHDU(syn_test)
hdul = fits.HDUList([hdu])
hdul.writeto('syn_test.fits',overwrite=True)'''
```

```python

ff_test = sim_test.gen_freefree_map()
'''hdu = fits.PrimaryHDU(ff_test)
hdul = fits.HDUList([hdu])
hdul.writeto('ff_test.fits',overwrite=True)'''
```

```python

egfg_test = sim_test.gen_exgal_map()
'''hdu = fits.PrimaryHDU(egfg_test)
hdul = fits.HDUList([hdu])
hdul.writeto('egfg_test.fits',overwrite=True)'''
```

```python
i = 0

fig = plt.figure(figsize=(12,8))

ax = fig.add_subplot(2,2,1)
ax.set_title('Synchrotron')
ax.set_xlabel('pixels')
im0 = ax.imshow(syn_test[:,:,i], cmap='inferno')
cbar = fig.colorbar(im0,ax=ax)
cbar.set_label('K', rotation=360)

ax = fig.add_subplot(2,2,2)
ax.set_title('Free-free')
ax.set_xlabel('pixels')
im0 = ax.imshow(ff_test[:,:,i], cmap='inferno')
cbar = fig.colorbar(im0,ax=ax)
cbar.set_label('K', rotation=360)

ax = fig.add_subplot(2,2,3)
ax.set_title('Extragalactic')
ax.set_xlabel('pixels')
im0 = ax.imshow((egfg_test[:,:,i]), cmap='inferno')
cbar = fig.colorbar(im0,ax=ax)
cbar.set_label('K', rotation=360)

ax = fig.add_subplot(2,2,4)
ax.set_title('SNR')
ax.set_xlabel('pixels')
im0 = ax.imshow((snr_test[:,:,i]), cmap='inferno')
cbar = fig.colorbar(im0,ax=ax)
cbar.set_label('K', rotation=360)

plt.tight_layout()
plt.show()
```
