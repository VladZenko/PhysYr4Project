```python
import cv2

in_path_fg = "./FastICA_Data/"
in_path_cs = "./FastICA_Data/"

output_path = "./FastICA_Data/"
fname_all = "test_filename"

fname_cs = "observer_lightcone"
fname_noise = "noise_108.000MHz_210.300MHz_SKA_SKA_central_area_EOR0_0128_3.0_1000h_K"


indata_cs = fits.getdata(in_path_cs+fname_cs+'.fits')
indata_cs = indata_cs/1000 # 21cmFAST are in mK by default, convert to K

indata_noise = fits.getdata(in_path_cs+fname_noise+'.fits')
indata_noise = indata_noise[:316, :, :]
arr = np.zeros((316,150,150))
for i in range(316):
    arr[i,:,:] = cv2.resize((indata_noise[i,:,:]-np.min(indata_noise[i,:,:])), (150, 150))+np.min(indata_noise[i,:,:])


indata_cs = indata_cs.transpose(2, 0, 1)
indata_noise = arr


rms_exgal = []
for i in range(316):
    rms_exgal.append(np.sqrt(np.mean(egfg_test[:,:,i]**2)))

rms_ff = []
for i in range(316):
    rms_ff.append(np.sqrt(np.mean(ff_test[:,:,i]**2)))

rms_syn = []
for i in range(316):
    rms_syn.append(np.sqrt(np.mean(syn_test[:,:,i]**2)))

rms_snr = []
for i in range(316):
    rms_snr.append(np.sqrt(np.mean(snr_test[:,:,i]**2)))

rms_noise = []
for i in range(316):
    rms_noise.append(np.sqrt(np.mean(indata_noise[i,:,:]**2)))

rms_cs = []
for i in range(316):
    rms_cs.append(np.sqrt(np.mean(indata_cs[i,:,:]**2)))

rms_fg = []
for i in range(316):
    rms_fg.append(np.sqrt(np.mean(tot_fg[:,:,i]**2)))
```

```python
nfreq=316
numin = 101.14547051514067
nustep = 0.17978510057201902

x = np.arange(numin, 315*nustep+numin, nustep)
ticks = np.arange(numin, 315*nustep+numin, 5)


plt.figure(figsize=(10,8))
plt.subplot(211)
plt.plot(x, rms_cs, linewidth=2, label='Cosmological signal')
plt.plot(x, rms_noise, linewidth=2, label='Noise')
plt.plot(x, np.array(rms_fg)*0.001,'--',color='purple', linewidth=2, label=r'Combined foregrounds $\times$ 0.001')
plt.ylabel('K',rotation=360)
plt.xlabel('Frequency, MHz')
plt.xticks(ticks[::-1], rotation=45)
plt.legend()
plt.subplot(212)
plt.plot(x, np.array(rms_exgal),'navy',linewidth=2, label=r'Extragalactic fg')
plt.plot(x, np.array(rms_ff),'magenta',linewidth=2, label=r'Free-free fg')
plt.plot(x, np.array(rms_syn)*0.1, 'red', linewidth=2, label=r'Synchrotron fg $\times$ 0.1')
plt.plot(x, np.array(rms_snr)*0.01, 'green', linewidth=2, label=r'SNR fg $\times$ 0.01')
plt.ylabel('K',rotation=360)
plt.xlabel('Frequency, MHz')
plt.xticks(ticks[::-1], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
