```python
lightcone = p21c.run_lightcone(
    redshift = 5.0,
    max_redshift = 12.0,
    user_params = {"HII_DIM":500, "BOX_LEN": 2000},
    lightcone_quantities=("brightness_temp", 'density'),
    global_quantities=("brightness_temp", 'density', 'xH_box'),
    direc='_cache'
)
```

```python
plotting.lightcone_sliceplot(lightcone);
```

```python
dir = './Boxes/'
fname = 'test'
fits.writeto(dir+fname+'.fits',np.transpose(lightcone.brightness_temp,(2,0,1)),overwrite=True)
```
