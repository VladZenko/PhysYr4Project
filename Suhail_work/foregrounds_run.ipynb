```python
from foreground_sim import ForeGsim
import astropy.io.fits as fits
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
sim_test = ForeGsim(nfreq=342,numin = 108, nustep = 0.3,ngrid = 128, imgsize=5.0) #342
```

```python
%%time
syn_test = sim_test.gen_sync_map()  #synchotron emission
hdu = fits.PrimaryHDU(syn_test)
hdul = fits.HDUList([hdu])
hdul.writeto('syn_test.fits',overwrite=True)
```

```python
plt.imshow(syn_test[:,:,0])
print(np.mean(syn_test[:,:,0]))
```

```python
%%time
ff_test = sim_test.gen_freefree_map() #free free emission
hdu = fits.PrimaryHDU(ff_test)
hdul = fits.HDUList([hdu])
hdul.writeto('ff_test.fits',overwrite=True)
```

```python
plt.imshow(ff_test[:,:,0])
print(np.mean(ff_test[:,:,0]))
```

```python
%%time
egfg_test = sim_test.gen_exgal_map() #extra galatic
hdu = fits.PrimaryHDU(egfg_test)
hdul = fits.HDUList([hdu])
hdul.writeto('egfg_test.fits',overwrite=True)
```

```python
plt.imshow(egfg_test[:,:,0])
print(np.mean(egfg_test[:,:,0]))
```

```python
new_egfg = np.zeros(np.shape(egfg_test))
#normalise and centre at zero
for i in range(342):
    x = np.mean(egfg_test[:,:,i])
    new_egfg[:,:,i] = (egfg_test[:,:,i]/x) - 1
print(np.mean(new_egfg[:,:,30]))
```

```python
#check if zero meaned
for i in range(342):
    x = np.mean(new_egfg[:,:,i])
    if x< -0.1 and x> 0.1:
        print('slice '+str(i)+' is not zero meaned')
        break
        os.system('clear')
        print('Is zero meaned')
```

```python
fg = syn_test + ff_test +new_egfg

fig = plt.figure(figsize=(24,6))

ax = fig.add_subplot(1,4,1)
im0 = ax.imshow(syn_test[:,:,0])
ax.set_title('synchotron emission')
fig.colorbar(im0,ax=ax)

ax = fig.add_subplot(1,4,2)
im0 = ax.imshow(ff_test[:,:,0])
ax.set_title('free free emission')
fig.colorbar(im0,ax=ax)

ax = fig.add_subplot(1,4,3)
im0 = ax.imshow(new_egfg[:,:,0])
ax.set_title('Galactic')
fig.colorbar(im0,ax=ax)

ax = fig.add_subplot(1,4,4)
im0 = ax.imshow(fg[:,:,0])
ax.set_title('Total foreground')
fig.colorbar(im0,ax=ax)
plt.tight_layout()
plt.show()

fig.savefig('Foreground.png')
```

```python
hdu = fits.PrimaryHDU(fg)
hdul = fits.HDUList([hdu])
hdul.writeto('fg_all.fits',overwrite=True)
```
